```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://covid.ourworldindata.org/data/ecdc/full_data.csv')
data
```

```python
def plot_new_cases(loc, ax=None):
    if ax is None:
        ax = plt.gca()
    data[data["location"] == loc].plot(x='date', y='new_cases', marker='.', ax=ax, label=loc)

ax = plt.gca()
plot_new_cases("Germany", ax=ax)
plot_new_cases("Netherlands", ax=ax)
plot_new_cases("Austria", ax=ax)

ax.set_ylabel('new cases');
```
